# Partitioning the data 75:25

```python
X = data[['Humidity3pm','Rainfall','RainToday']]
Y = data[['RainTomorrow']]
```

# logistic Regression

```python

from  sklearn.preprocessing import MinMaxScaler
minmax = MinMaxScaler(feature_range = (0,1))
minmax.fit(X).transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.25,random_state=1)
X_train_std=minmax.fit_transform(X_train)
X_test_std=minmax.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score,cross_val_predict
lr=LogisticRegression()
lr.fit(X_train,Y_train)
lr_acc=cross_val_score(lr,X_train_std,Y_train,cv=3,scoring='accuracy',n_jobs=-1)
lr_proba=cross_val_predict(lr,X_train_std,Y_train,cv=3,method='predict_proba')
lr_scores=lr_proba[:,1]
```

```python
lr_acc
```

```python
#predict on test data
Y_pred=lr.predict(X_test)
```

```python
print('Actual values')
print(Y_test)
```

```python
print("Predicted values ")
print(Y_pred)
```

```python
from sklearn.metrics import accuracy_score,recall_score,roc_auc_score,confusion_matrix
print("Recall score %f" %(recall_score(Y_test,Y_pred)*100))
print("ROC_AUC score %f" %(roc_auc_score(Y_test,Y_pred)*100))
lrcm=confusion_matrix(Y_test,Y_pred)
lrcm
```

```python
print("Sensitivity_lrcm/Recall_lrcm ",lrcm[0,0]/(lrcm[0,0]+lrcm[1,0]))
print("Specificity_lrcm ",lrcm[1,1]/(lrcm[1,1]+lrcm[0,1]))
print("Precision_lrcm ",lrcm[0,0]/(lrcm[0,0]+lrcm[0,1]))
```

```python
from sklearn.metrics import roc_auc_score,roc_curve
def ROC_curve(title,Y_train,scores,label=None):
#caluculate the roc score
    fpr,tpr,thresholds=roc_curve(Y_train,scores)
    print('AUC Score ({}):{:2f}'.format(title,roc_auc_score(Y_train,scores)))
    
    plt.figure(figsize=(8,6))
    plt.plot(fpr,tpr,linewidth=2,label=label,color='b')
    plt.xlabel('False positive Rate',fontsize=16)
    plt.ylabel('True positive Rate',fontsize=16)
    plt.title('ROC Curve:{}'.format(title),fontsize=16)
    plt.show()
```

```python

ROC_curve('Logistic Regression',Y_train,lr_scores)
ROC_curve('Decision Tree Classification',Y_train,dtc_clf_scores)
ROC_curve('Random Forest clssification',Y_train,rmf_clf_scores)
```
